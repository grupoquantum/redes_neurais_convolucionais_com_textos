<h3>Redes Neurais Convolucionais na Classificação de Textos</h3>

<h4>Instanciação e Treinamento</h4>

```python
from Neuraline.ArtificialIntelligence.DeepLearning.convolutional_neural_network import ConvolutionalNeuralNetwork
from Neuraline.Utilities.chart import Chart
convolutional_neural_network, chart = ConvolutionalNeuralNetwork(preprocessor='standardization'), Chart()

convolutional_neural_network.addConvolutionalLayer(feature_maps=[1000, 1000], max_pooling=False, number_of_cores=100, step=1, activation_function='nonlinear')

result1 = convolutional_neural_network.addFit(url_path='./texts/training/english', class_name='inglês')
result2 = convolutional_neural_network.addFit(url_path='./texts/training/portuguese', class_name='português')
if result1 and result2: print('treinamento concluído com sucesso')
else: print('erro ao concluir o treinamento')
```

<h4>Teste com os Textos em Inglês</h4>

```python
url_inputs = './texts/test/english.txt'
chart.plotTEXT(url_inputs)
classification = convolutional_neural_network.predict(url_inputs)
print(f'este texto está em {classification}')
```

<h4>Teste com os Textos em Português</h4>

```python
url_inputs = './texts/test/portuguese.txt'
chart.plotTEXT(url_inputs)
classification = convolutional_neural_network.predict(url_inputs)
print(f'este texto está em {classification}')
```
